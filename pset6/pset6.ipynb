```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# Plot Bode Plots

bode1 = pd.read_csv('bode_plot_1.csv')
bode1.plot('Frequency (Hz)', 'Channel 2 Magnitude (dB)', title='Circuit 1')

bode2 = pd.read_csv('bode_plot_2.csv')
bode2.plot('Frequency (Hz)', 'Channel 2 Magnitude (dB)', title='Circuit 2')

bode3 = pd.read_csv('bode_plot_3.csv')
bode3.plot('Frequency (Hz)', 'Channel 2 Magnitude (dB)', title='Circuit 3')

bode4 = pd.read_csv('bode_plot_4.csv')
bode4.plot('Frequency (Hz)', 'Channel 2 Magnitude (dB)', title='Circuit 4')

bode5 = pd.read_csv('bode_plot_5.csv')
bode5.plot('Frequency (Hz)', 'Channel 2 Magnitude (dB)', title='Circuit 5')
```

```python
# Calculate analytical bode plot for circuit 1
R = 1.58 * 10 ** 3  # ohm
C = 0.1 * 10 ** -6  # Farad
tau = R * C

omega = bode1['Frequency (Hz)'] * 2 * np.pi

A = 20 * np.log10(1 / (1 + (R * C * omega) ** 2))
bode1['A_theory'] = A
bode1.plot('Frequency (Hz)', ['A_theory', 'Channel 2 Magnitude (dB)'], title='Circuit 1')
```
